### Q1:計算標準常態分配，小於1的機率有多大?

### Q2:計算標準常態分配，大於1，小於 -1 的機率有多大?
![](Q2.png)


```python
## Q2
'''
你的程式碼
'''
mu=0
sigma=1
cumsum_probs = 1 - stats.norm.cdf(1, mu, sigma) + stats.norm.cdf(-1, mu, sigma)
cumsum_probs
```
