```python
# library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
```

```python
#輸入資料
boys=[164, 176, 169, 169, 165, 175, 159, 151, 144, 160, 183, 165, 156, 170,
 164, 173, 165, 163, 177, 171]

girds=[169, 170, 162, 154, 183, 173, 169, 167, 170, 185, 162, 175, 168,
  151, 181, 170, 182, 156, 159, 160]
```

```python
#Q1:試著用今天所教的內容，如何描述這兩組資料的樣態?
#暗示:  分別計算男生和女生的統計量
'''
你的答案
'''
print("--- boys ---")
print("mean=",statistics.mean(boys))
print("median=",statistics.median(boys))
print("mode=",statistics.mode(boys))
print("variance=",statistics.variance(boys))
print("std=",statistics.stdev(boys))
#最後，畫圖看分布
plt.hist(boys,alpha=.4,bins=40)
plt.title('boy,skewness={0},kurtosis={1}'.format(round(stats.skew(boys),2),round(stats.kurtosis(boys),2)))
plt.axvline(x=statistics.mean(boys))
plt.show()


print("\n")
print("--- girds ---")
print("mean=",statistics.mean(girds))
print("median=",statistics.median(girds))
print("mode=",statistics.mode(girds))
print("variance=",statistics.variance(girds))
print("std=",statistics.stdev(girds))
#最後，畫圖看分布
plt.hist(girds,alpha=.4,bins=40)
plt.title('girds,skewness={0},kurtosis={1}'.format(round(stats.skew(girds),2),round(stats.kurtosis(girds),2)))
plt.axvline(x=statistics.mean(girds))
plt.show()
```

```python
#Q2: 請問男生和女生在平均身高上誰比較高? 
'''
你的答案 -> 女生
'''
print("statistics_mean_boy=",statistics.mean(boys))
print("statistics_mean_gird=",statistics.mean(girds))
```
